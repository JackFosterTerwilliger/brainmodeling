```python
def plotRate(spikes, second, smoothing_window, step):
    rate = []
    for i in range(int(smoothing_window/2.) , len(spikes) - int(smoothing_window/2.), step):
        rate.append(  sum(spikes[ int(i-(smoothing_window/2.)):int(i+(smoothing_window/2.))])/10. )
    return rate
```

```python
def smoothListGaussian(list,degree=5):  

     window=degree*2-1  

     weight=np.array([1.0]*window)  

     weightGauss=[]  

     for i in range(window):  

         i=i-degree+1  

         frac=i/float(window)  

         gauss=1/(np.exp((4*(frac))**2))  

         weightGauss.append(gauss)  

     weight=np.array(weightGauss)*weight  

     smoothed=[0.0]*(len(list)-window)  

     for i in range(len(smoothed)):  

         smoothed[i]=sum(np.array(list[i:i+window])*weight)/sum(weight)  

     return smoothed
```

```python
def cV(spikes):
    interspike_interval=[]
    for i in range(len(spikes)-2):
        interspike_interval.append(spikes[i+1] - spikes[i])
    mean=sum(interspike_interval)/len(interspike_interval)
    sd=np.std(interspike_interval)
    return sd/mean
```

<center>$\LARGE\text{Integrate and Fire Neuron with STDP}$</center>

$\large\text{Neuron model}$

$\tau_m \frac{dV}{dt} = V_{\text{rest}} - V + g_{ex}(t)(E_{ex} - V)  + g_{in}(t)(E_{in} - V)\\$

$\large\text{Peak Conductance}$

$\large\tau_{\text{ex}} \frac{dg_{\text{ex}}}{dt} = -g_{\text{ex}}$

$\large\tau_{\text{in}} \frac{dg_{\text{in}}}{dt} = -g_{\text{in}}$

\begin{align}
\text{if} \quad \text{presynaptic ex neuron fires} 
\quad &\text{then  } g_{ex} \leftarrow g_{ex} + \bar{g}_a\\
\text{if} \quad \text{presynaptic in neuron fires} 
\quad &\text{then  } g_{in} \leftarrow g_{in} + \bar{g}_{in}
\end{align}


$\large\text{STDP}$

$\large\tau_{-} \frac{dM}{dt} = -M$

$\large\tau_{+} \frac{dP_a}{dt} = -P_a$

\begin{align}
\text{if} \quad \text{postsynaptic ex neuron fires} 
\quad &\text{then} \quad \Huge\{ \large \begin{array}{l} M \leftarrow A_- + M \\\\ g_{ex} \leftarrow g_{ex} + P_ag_{max} \end{array}\\
\text{if} \quad \text{presynaptic in neuron fires} 
\quad &\text{then} \quad \Huge\{ \large \begin{array}{l} M \leftarrow A_+ + P_a \\\\ g_{in} \leftarrow g_{in} + Mg_{max} \end{array}
\end{align}

$A_+ = .005 $

$A_-/A_+ = 1.05$

$\tau_- = \tau_+ = 20 \text{msec} $

```python
#Plotting the distribution of synaptic weights

#Time
dt = .5
msec = 1.0/dt
sec = 1000.0*msec
time = 200*sec #simulation duration

#Neuron Parameters
TAU = 20.0
V_REST = -70.0
V_RESET = -60.0
V_TH= -54.0

#Connection Parameters
TAU_EX = 5.0
TAU_IN = 5.0
E_EX = 0.0
E_IN = -70.0

#STDP parameters
TAU_P = 20.0
TAU_M = 20.0
A_pos = .005
A_neg = A_pos * 1.05

#Network Parameters
N = 1
EX = 100
IN = 20

#Presynaptic Firing Frequencies
freq = [10,15,20,25,30,40]

#Plotting Parameters
fig, axes = plt.subplots(nrows=len(freq)/2, ncols=2)
fig.set_size_inches(15,15)

for fi in range(len(freq)):

    v = -65.0 #initial membrane potential
    g_ex = np.ones(EX)*0.0 #initial excitatory conductances
    g_in = np.ones(IN)*0.0 #initial inhibitory conductances
    ga = np.ones(EX)*0.15 #initial excitatory weights are maxed out
    gmax = np.ones(EX)*0.15 #maximum conductance value
    gi = np.ones(IN)*0.5 #constant inhibitory weights are uniform
    zero = np.zeros(EX)
    p=np.ones(EX)*0.0 #amount to boost excitatory weights
    m=np.ones(EX)*0.0 #amount to decriment excitatory weights


    spiketrain_ex = rndm.randint(1,int(sec/freq[fi]),EX) #random distribution of initial presynaptic action potential
    spiketrain_in = rndm.randint(1,int(sec/10),IN) #random distribution of initial presynaptic action potential

    spikes=[]
    for t in np.arange(0, time,1):

        ex_spikes = np.where(spiketrain_ex <= 0)[0] #get the indicies of firing presynaptic excitatory neurons
        noex_spikes = np.where(spiketrain_ex > 0)[0] #get the indicies of non-firing presynaptic excitatory neurons
        in_spikes = np.where(spiketrain_in <= 0)[0]  #get the indicies of firing presynaptic inhibitory neurons
        noin_spikes = np.where(spiketrain_in > 0)[0] #get the indicies of non-firing presynaptic inhibitory neurons

        #presynaptic excitatory neurons
        g_ex[noex_spikes] += -g_ex[noex_spikes]/TAU_EX*dt #decay conductance
        p -= p/TAU_P*dt #decay weight-increase value
        if len(ex_spikes)>0: #if the pre-synaptic neuron fires
            g_ex[ex_spikes] += ga[ex_spikes] #increment conductance
            p[ex_spikes] += A_pos #increment weight-increase value
            ga[ex_spikes] += m[ex_spikes]*gmax[ex_spikes] #decriment the synaptic weight
            spiketrain_ex[ex_spikes] = rndm.poisson(sec/freq[fi],EX)[ex_spikes] #time of next AP

        #presynaptic inhibitory neurons
        g_in[noin_spikes] += -g_in[noin_spikes]/TAU_IN*dt #decay conductance
        if len(in_spikes)>0: #if the pre-synaptic neuron fires
            g_in[in_spikes] += gi[in_spikes] #increment the synaptic weight
            spiketrain_in[in_spikes] = rndm.poisson(sec/10,IN)[in_spikes] #time of next AP

        m -= m/TAU_M*dt #decriment weight-decrease value
        #membrane potential
        if v >= V_TH: #if the post-synaptic neuron fires
            ga += p*gmax #increase the synaptic weight
            m += -A_neg #increase the weight-decrease value
            v = V_RESET #reset the post-synaptic neuron's membrane potential
            spikes.append(1)
        else:
            spikes.append(0)

        I = (g_ex*(E_EX - v)).sum() + (g_in*(E_IN - v)).sum() #calculate input to the post-synaptic neuron
        v += (V_REST - v + I)/TAU*dt #compute the post-synaptic neuron's membrane potential

        #keep synaptic weights in their bounds
        ga = np.minimum(ga,gmax)
        ga = np.maximum(ga,zero)
        #update time-till next presynaptic AP
        spiketrain_ex -= 1
        spiketrain_in -= 1
    
    #Plot the % of each synaptic weight
    i=fi/2
    j=fi%2
    axes[i][j].set_title("Distribution of Peak Conductances at " + str(freq[fi]) + " Herz")
    axes[i][j].set_ylabel("Fraction in Bin")
    axes[i][j].set_xlabel("g/gmax")
    weights = np.ones_like(ga)/len(ga)
    axes[i][j].grid(b=True, which='major', color='k', linestyle=':')
    axes[i][j].hist(ga/gmax, 20,weights=weights)
plt.tight_layout()
```

```python
#Plotting the distribution of synaptic weights

#Time
dt = .5
msec = 1.0/dt
sec = 1000.0*msec
time = 200*sec #simulation duration

#Neuron Parameters
a = 0.02
b = 0.2
c = -65.0
d = 8.0

#Connection Parameters
TAU_EX = 5.0
TAU_IN = 5.0
E_EX = 0.0
E_IN = -70.0

#STDP parameters
TAU_P = 20.0
TAU_M = 20.0
A_pos = .01
A_neg = A_pos * 1.05

#Network Parameters
N = 1
EX = 100
IN = 20

#Presynaptic Firing Frequencies
freq = [10,15,20,25,30,40]

#for plotting
fig, axes = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(15,15)
cv=[]
frequency=[]

for fi in range(len(freq)):

    v = -65.0 #initial membrane potential
    u=v*b
    g_ex = np.ones(EX)*0.0 #initial excitatory conductances
    g_in = np.ones(IN)*0.0 #initial inhibitory conductances
    ga = np.ones(EX)*0.15 #initial excitatory weights are maxed out
    gmax = np.ones(EX)*0.15 #maximum conductance value
    gi = np.ones(IN)*0.5 #constant inhibitory weights are uniform
    zero = np.zeros(EX)
    p=np.ones(EX)*0.0 #amount to boost excitatory weights
    m=np.ones(EX)*0.0 #amount to decriment excitatory weights


    spiketrain_ex = rndm.randint(1,int(sec/freq[fi]),EX) #random distribution of initial presynaptic action potential
    spiketrain_in = rndm.randint(1,int(sec/10),IN) #random distribution of initial presynaptic action potential

    spikes=[]
    interspikes=[]
    for t in np.arange(0, time,1):

        ex_spikes = np.where(spiketrain_ex <= 0)[0] #get the indicies of firing presynaptic excitatory neurons
        noex_spikes = np.where(spiketrain_ex > 0)[0] #get the indicies of non-firing presynaptic excitatory neurons
        in_spikes = np.where(spiketrain_in <= 0)[0]  #get the indicies of firing presynaptic inhibitory neurons
        noin_spikes = np.where(spiketrain_in > 0)[0] #get the indicies of non-firing presynaptic inhibitory neurons

        #presynaptic excitatory neurons
        g_ex[noex_spikes] += -g_ex[noex_spikes]/TAU_EX*dt #decay conductance
        p -= p/TAU_P*dt #decay weight-increase value
        if len(ex_spikes)>0: #if the pre-synaptic neuron fires
            g_ex[ex_spikes] += ga[ex_spikes] #increment conductance
            p[ex_spikes] += A_pos #increment weight-increase value
            ga[ex_spikes] += m[ex_spikes]*gmax[ex_spikes] #decriment the synaptic weight
            spiketrain_ex[ex_spikes] = rndm.poisson(sec/freq[fi],EX)[ex_spikes] #time of next AP

        #presynaptic inhibitory neurons
        g_in[noin_spikes] += -g_in[noin_spikes]/TAU_IN*dt #decay conductance
        if len(in_spikes)>0: #if the pre-synaptic neuron fires
            g_in[in_spikes] += gi[in_spikes] #increment the synaptic weight
            spiketrain_in[in_spikes] = rndm.poisson(sec/10,IN)[in_spikes] #time of next AP

        m -= m/TAU_M*dt #decriment weight-decrease value
        #membrane potential
        if v >= 30: #if the post-synaptic neuron fires
            ga += p*gmax #increase the synaptic weight
            m += -A_neg #increase the weight-decrease value
            v = c #reset the post-synaptic neuron's membrane potential
            u = u + d
            interspikes.append(t)
            spikes.append(1)
        else:
            spikes.append(0)

        I = (g_ex*(E_EX - v)).sum() + (g_in*(E_IN - v)).sum() #calculate input to the post-synaptic neuron
        v += dt*(0.04*v**2 + 5*v + 140 - u + I) #compute the post-synaptic neuron's membrane potential
        u += dt*(a*(b*v - u))

        #keep synaptic weights in their bounds
        ga = np.minimum(ga,gmax)
        ga = np.maximum(ga,zero)
        #update time-till next presynaptic AP
        spiketrain_ex -= 1
        spiketrain_in -= 1
    
    #Plot the % of each synaptic weight
    cv.append(cV(interspikes))
    frequency.append( sum(spikes[int(len(spikes)-sec):])/10 )
axes[0].set_title("Frequency of Postsynaptic Firing Rate given Input Rates")
axes[0].set_ylabel("Postsynaptic Frequency (Hz)")
axes[0].set_ylim((0,20))
axes[0].set_xlabel("Presynaptic Frequency (Hz)")
axes[0].grid(b=True, which='major', color='k', linestyle=':')
axes[0].plot(freq,frequency)

axes[1].set_title("CV of Postsynaptic Firing Rate given Input Rates")
axes[1].set_ylabel("Coefficient of Variation")
axes[1].set_ylim((0,1))
axes[1].set_xlabel("Presynaptic Frequency (Hz)")
axes[1].grid(b=True, which='major', color='k', linestyle=':')
axes[1].plot(freq,cv)

plt.tight_layout()
```
