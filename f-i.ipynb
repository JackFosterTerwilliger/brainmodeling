$v' = 0.04v^2 + 5v + 140 - u + I$

$u' = a(bv - u)$

\begin{equation}
\text{if} \quad v \ge +30 \text{,} 
\quad \text{then} \quad \{ \begin{array}{l} v \leftarrow c \\\\ u \leftarrow u + d \end{array}
\end{equation}

```python
def dirac(x):
    if x == 0:
        return 1
    return 0
```

```python
class ExcitatoryConnections:
    AMPA_TAU = 5
    NMDA_TAU = 150
    T_pos=15
    T_neg=20
    A_pos=.004
    A_neg=.004
    
    U=0.5
    F=1000
    D=800
        
    def __init__(self, weights, ratio=0, ifstdp=False, ifstd=False):
        self.w = weights
        self.ratio = ratio
        self.ampa_g = np.zeros(weights.shape)
        self.nmda_g = np.zeros(weights.shape)
        self.ifstdp=ifstdp
        self.ifstd=ifstd
        
        self.x=1
        self.u=self.U
        
    def step(self, last_spikes, fired, t):
        self.ampa_g += -self.ampa_g/self.AMPA_TAU
        self.nmda_g += -self.nmda_g/self.NMDA_TAU

        self.ampa_g[:,fired] += self.w[:,fired]*self.x*self.u
        self.nmda_g[:,fired] += self.ratio*self.w[:,fired]*self.x*self.u

        if self.ifstdp:
            self.stdp(last_spikes,t)
        
        if self.ifstd:
            self.std(last_spikes,t)
    
    def stdp(self, last_spikes, t):
        for i_post in range(len(self.w)):
            
            for i_pre in range(len(self.w[i_post])):
                if last_spikes[i_pre] == t-1 or last_spikes[i_post] == t-1:
                    
                    if last_spikes[i_post] >=0 and last_spikes[i_pre] >= 0:
                        t_diff = last_spikes[i_post] - last_spikes[i_pre]

                        if t_diff > 0: #if post synaptic spike is correlated with this connection 
                            self.w[i_post,i_pre] = self.w[i_post,i_pre] + self.A_pos*np.exp(-t_diff/self.T_pos) #increment the weight
                            if self.w[i_post,i_pre] >= .5:
                                self.w[i_post,i_pre] = .5
                        else: #if post synaptic spike occurs before the 
                            self.w[i_post,i_pre] += - self.A_neg*np.exp(t_diff/self.T_neg) #deecrement the weight
                            if self.w[i_post,i_pre] < 0:
                                self.w[i_post,i_pre] = 0
                                
    def std(self, last_spikes, t):
        for post in self.w:
            for i_pre in range(len(post)):
                if last_spikes[i_pre] != None:
                    self.x = self.x + (1 - self.x)/self.D - self.u*self.x*dirac(t-1 - last_spikes[i_pre])
                    self.u = self.u + (self.U-self.u)/self.F + self.U*(1 - self.u)*dirac(t-1 - last_spikes[i_pre])

    def getSynapticCurrent(self, v):
        return self.ampa_g.sum(1)*self.ampa(v) + self.nmda_g.sum(1)*self.nmda(v)
    
    def ampa(self, v):
        return v
    
    def nmda(self, v):
        return v*(((v + 80)/60)**2/(1 + ((v + 80)/60)**2))
```

```python
class InhibatoryConnections:
    GABA_A_TAU = 6
    GABA_B_TAU = 100
    
    U = 0.2
    F = 20
    D = 700

    def __init__(self, weights, ratio=0, ifstd=False):
        self.w = weights
        self.ratio = ratio
        self.gaba_a_g = np.zeros(weights.shape)
        self.gaba_b_g = np.zeros(weights.shape)
        self.ifstd=ifstd
        self.x=1
        self.u=self.U
        
    def step(self, last_spikes, fired, t):
        self.gaba_a_g += -self.gaba_a_g/self.GABA_A_TAU
        self.gaba_b_g += -self.gaba_b_g/self.GABA_B_TAU

        self.gaba_a_g[:,fired] += self.w[:,fired]*self.x*self.u
        self.gaba_b_g[:,fired] += self.ratio*self.w[:,fired]*self.x*self.u
        
        if self.ifstd:
            self.std(last_spikes, t)
            
    def std(self, last_spikes, t):
        for post in self.w:
            for i_pre in range(len(post)):
                if last_spikes[i_pre] != None:
                    self.x = self.x + (1 - self.x)/self.D - self.u*self.x*dirac(t-1 - last_spikes[i_pre])
                    self.u = self.u + (self.U-self.u)/self.F + self.U*(1 - self.u)*dirac(t-1 - last_spikes[i_pre])


    def getSynapticCurrent(self, v):
        return self.gaba_a_g.sum(1)*self.gaba_a(v) + self.gaba_b_g.sum(1)*self.gaba_b(v)
    
    def gaba_a(self,v):
        return v + 70
    
    def gaba_b(self, v):
        return v + 90
```

```python
class PoissonInput:
    
    def __init__(self, rate, size, func):
        self.spikegen = func
        self.rate = rate
        self.size = size
        self.poisson = self.spikegen(self.rate,self.size)
    
    def step(self):
        spiking = np.where(self.poisson <= 0)[0]
        self.poisson[spiking] = self.spikegen(self.rate,self.size)[spiking]
        self.poisson -= 1
        return spiking
```

```python
def fpoisson(rate,size):
    return np.random.poisson(rate,size)
```
